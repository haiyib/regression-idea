# Clean EH Data Portal Obesity CSV\nFilter to borough-level obesity rates for the five NYC boroughs.

```python
import pandas as pd

df = pd.read_csv("NYC EH Data Portal - Obesity (adults) (full table).csv")
df.head()
```

```python
# Filter to borough-level rows only
borough_df = df[df["GeoType"] == "Borough"].copy()

# Extract the point estimate from the Percent column (e.g. "37.4 (33.8, 41.1)" -> 37.4)
borough_df["Obesity_Rate"] = borough_df["Percent"].str.extract(r"^([\d.]+)").astype(float)

# Keep only the columns we need and rename
clean_df = borough_df[["TimePeriod", "Geography", "Obesity_Rate"]].rename(columns={
    "TimePeriod": "Year",
    "Geography": "Borough"
})

clean_df = clean_df.sort_values(["Year", "Borough"]).reset_index(drop=True)
clean_df
```

```python
# Save to CSV
clean_df.to_csv("borough_obesity_rates.csv", index=False)
print(f"Saved {len(clean_df)} rows to borough_obesity_rates.csv")
```

```python
import pandas as pd

df = pd.read_csv('raw-census-tract-data.csv')
pd.set_option('display.max_columns', None)  # Show all columns
print(df.head())
```

```python
df_ny = df[df['StateAbbr'] == 'NY'].sort_values(['StateAbbr', 'OBESITY_CrudePrev'])
```

```python
print(df_ny.head())
```

```python
df_ny = df[df['StateAbbr'] == 'NY'][['StateAbbr', 'StateDesc', 'CountyName', 'CountyFIPS', 'TractFIPS', 'OBESITY_CrudePrev']].sort_values('OBESITY_CrudePrev', ascending=False)
print(df_ny.head())
```

```python
df_ny.to_csv('ny_obesity.csv', index=False)
```
