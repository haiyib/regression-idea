# Merge NY Obesity CSVs (2023, 2024, 2025)
Concatenate the three yearly tract-level obesity files into one, tagging each row with its year.

```python
import pandas as pd

files = {
    2023: "ny_obesity_2023.csv",
    2024: "ny_obesity_2024.csv",
    2025: "ny_obesity_2025.csv",
}

frames = []
for year, path in files.items():
    df = pd.read_csv(path)
    df["Year"] = year
    frames.append(df)

merged = pd.concat(frames, ignore_index=True)
merged.head()
```

```python
merged.to_csv("ny_obesity_merged.csv", index=False)
print(f"Saved {len(merged)} rows to ny_obesity_merged.csv")
```
